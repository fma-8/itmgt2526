```python
import requests
import matplotlib.pyplot as plt
import time
from datetime import datetime
```

```python
api = 'https://api.wheretheiss.at/v1/satellites/25544'

xvalues = []
yvalues = []
timestamp = []

while len(xvalues) != 100:
    response = requests.get(api)
    data = response.json()
    timestamp.append(datetime.fromtimestamp(data['timestamp']))
    xvalues.append(data['latitude'])
    yvalues.append(data['longitude'])
    time.sleep(10)

plt.figure(figsize=(15, 9))
plt.plot(xvalues, yvalues, marker = 'o', linestyle = '-', color ='gray', markerfacecolor = 'red', markersize = 5, linewidth = 3)
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("ISS Location")
plt.grid(True)

for i in range(len(timestamp)):
    plt.annotate(timestamp[i], (xvalues[i], yvalues[i]), textcoords='offset points', xytext=(4,0), ha='left', fontsize = 5)

plt.show()
```
